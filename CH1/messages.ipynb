```python
import os
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
from langchain_openai import ChatOpenAI
from load_dotenv import load_dotenv
load_dotenv()

if os.environ.get("OPENAI_API_KEY"):
  print("API Key exists")
else:
  raise ValueError("No API Key found")
```

```python
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.9)
messages=[SystemMessage(content="You are a genz assistant who always answers in fun way"),
          HumanMessage(content="Tell me a fun fact")]
llm.invoke(messages).content
```

```python
from langchain_core.prompts import PromptTemplate
llm=ChatOpenAI(model="gpt-4o-mini",temperature=0.9)
user_input=input("Enter a topic for fun fact:")
dynamic_prompt = PromptTemplate.from_template("Write a fun fact about the {topic}")
ready_prompt = dynamic_prompt.invoke({"topic":user_input})
llm.invoke(ready_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt_template = ChatPromptTemplate.from_messages([
  ("system","You are a Polite assistant"),
  ("user","write a fun fact about {topic}")
])
ready_prompt=prompt_template.invoke({"topic":"AI"})
ready_prompt.messages
llm.invoke(ready_prompt).content
```

```python
from langchain_core.prompts import ChatPromptTemplate
prompt_template = ChatPromptTemplate([
  ("system","You are a {tone} assistant"),
  ("user","write a fun fact on {topic}")
])
user_topic = input("enter a topic:")
user_tone = input("enter a tone:")
ready_prompt = prompt_template.invoke({"topic":user_topic,"tone":user_tone})
ready_prompt
llm.invoke(ready_prompt).content
```
