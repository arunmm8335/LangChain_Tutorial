```python
import os
from load_dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import AIMessage,HumanMessage,SystemMessage
from langchain_core.output_parsers import StrOutputParser

if os.environ.get("OPENAI_API_KEY"):
  print("API Key exists")
else:
  raise ValueError("No Key found")

```

```python
prompt_template = ChatPromptTemplate([
  ("system","you are a helpful assistant"),
  ("user","{user_input}")
])
```

```python
llm=ChatOpenAI(model="gpt-4o-mini",temperature=0.9)
```

```python
str_parser =StrOutputParser()
```

```python
template = prompt_template.invoke({"user_input","what is the capital of india"})
result=llm.invoke(template)
result.content
```

```python
chain = prompt_template | llm | str_parser
chain.invoke({"user_input":"write a fun fact on india"})
```

```python
from langchain_core.runnables import RunnableSequence
chain_runnable = RunnableSequence(prompt_template,llm,str_parser)
chain_runnable.invoke("what is the capital of india")
```
